### Inter Quantile Range Strategy 
- Outlier Detection
- Outlier Treatment

```python
#CodeSection9
# Use quantile function
# Find Q1 and Q3

Q1 = train['Annual_Premium'].quantile(0.25)
Q3 = train['Annual_Premium'].quantile(0.75)

#Find IQR 
IQR = Q3 - Q1

#Find Lower Limit and Higher Limit
Upper_Whisker = Q3 + 1.5*IQR
Lower_Whisker = Q1 - 1.5*IQR

for dataset in all_data:
  dataset['Annual_Premium'] = dataset['Annual_Premium'].apply(lambda x: x if x<Upper_Whisker else Upper_Whisker)
  dataset['Annual_Premium'] = dataset['Annual_Premium'].apply(lambda x: x if x>Lower_Whisker else Lower_Whisker)
```

## 8/ Prepare Data for Model Model

```python
#CodeSection16
# Check First 5 Rows of Test Data
test.head()
```

```python
train.head()
```

```python
#CodeSection17
# Get X and y Variables for your Dataset
X = train.drop(['id','Response'], axis =  1)
y = train['Response']
```

### Create Dummy Variables for  Categorical Variables

```python
#CodeSection18
# Get Dummy Variables
X = pd.get_dummies(X)  # Keep only n - 1 Variables - Use "drop_first = True"
```

## 9/ Build Model

```python
#CodeSection19
#Import Logistic Regression Library
from sklearn.linear_model import LogisticRegression

#Initialize LR
clf = LogisticRegression(max_iter = 1000)
```

```python
#CodeSection20
# Fit Model
model = clf.fit(X, y)
```

### Check Accuracy of Model on Train Data

```python
#CodeSection21
# Predict on Train Data
y_pred = model.predict(X)
```
